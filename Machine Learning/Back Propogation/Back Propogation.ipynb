```python
y = np.array([20,40,401.13,601.19,799.28,584.84,1156.96,901.36,1126.31,1280.4])
x = np.array([  1,  21,  41,  61,  81, 101, 121, 141, 161, 181])
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(),tf.keras.layers.Dense(1)])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# epochs is the hightration, how many times it is cmg back to check the weights/ errors
```

```python
model.fit(x, y, epochs=250)
```

```python
model.predict(np.array([15]))
```
